## another way of representing graph

```python
graph = {} # key will be current batch and value will be list of parents
unq_nodes = set() # only for development to have unique values

def build_graph(batch_id):
    parent_batch_list_from_db = []

    # build dummy parents
    for i in range(random.randint(0,2)):
        # find a unique batch
        d = random.randint(1,100)
        while d in unq_nodes:
            d = random.randint(1,100)

        parent_batch_list_from_db.append(str(d))
        unq_nodes.add(d)

    # remove None and current batch
    parent_batch_list_from_db = [batch for batch in parent_batch_list_from_db if (batch and batch != batch_id)]

    # store parent in graph
    graph[batch_id] = parent_batch_list_from_db

    # Build graph for each parent
    for parent in parent_batch_list_from_db:
        build_graph(parent)
```

```python
build_graph('111')
graph
```

```python
def show_graph(batch_id):
    print("current batch id - ", batch_id)
    print("Number of parents - ",len(graph[batch_id]))
    print("parents - ", graph[batch_id])
    print('-'*10)

    for parent in graph[batch_id]:
        show_graph(parent)

show_graph('111')
    
```

```python
from pyvis.network import Network
import json

net = Network(height="900px", width="100%", directed=True)

edges = []
for batch_id, parent_list in graph.items():
    for parent in parent_list:
        edges.append((batch_id, parent))
print(edges)

#  Add unique nodes first
all_nodes = set()
for parent, child in edges:
    all_nodes.add(parent)
    all_nodes.add(child)

for node in all_nodes:
    net.add_node(node, label=node)

# Now add edges
for parent, child in edges:
    net.add_edge(parent, child)

options = {
    "layout": {
        "hierarchical": {
            "enabled": True,
            "direction": "UD",
            "sortMethod": "directed",
            "levelSeparation": 200,
            "nodeSpacing": 250,
            "treeSpacing": 300
        }
    },
    "physics": {
        "enabled": False
    },
    "nodes": {
        "shape": "box",
        "font": {"size": 18}
    }
}

net.set_options(json.dumps(options))
net.write_html("graph.html")

print("open 'graph.html' in browser")
```
